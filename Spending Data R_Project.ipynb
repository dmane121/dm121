```r
ggplot(aes(Gender), data = df) +
  geom_bar(color = "black", fill = "brown")
```

It shows that Female customers are than male customer in shop for shopping.

```r
hist(df$Family.Size, ,main="Family.Size of customer",xlab="Family.Size",freq=TRUE,col="Green",border="Blue")
```

from above histogram we can conclude that customer whos family contain more than 7 person ,they go for shoping very rarely


```r
hist(df$Annual.Income...., ,main="Annual income",xlab="Annual.Income....",freq=TRUE,col="pink",border="Blue")
```

from above histogram we can conclude that customer whos Annual income in between 50000 to 100000 go to shop frequently.




```r
 ggplot(df, aes(x=Profession,y=Work.Experience,fill=Profession))+
  geom_boxplot()
```

```r
 ggplot(df, aes(x=Profession,y=Annual.Income....,fill=Profession))+
  geom_boxplot(width=0.5,position = position_dodge(width=50))+
    theme(axis.text.x=element_text(size=10),axis.text.y=element_text(size=15))+
  theme(axis.title.x=element_text(size=15),axis.title.y=element_text(size=15))
 
```

```r
ggplot(df, aes(x=Age, y=Annual.Income...., col=Annual.Income....))+
  geom_point()+
   geom_smooth(method="lm")+
     theme(axis.text.x=element_text(size=15),axis.text.y=element_text(size=15))+
  theme(axis.title.x=element_text(size=15),axis.title.y=element_text(size=15))
```

```r
ggplot(df, aes(Spending.Score..1.100., Age )) +
  geom_point(col="pink", size=4)+
  geom_smooth(method="lm",col="firebrick")+
  facet_wrap(~Gender)+  theme(axis.text.x=element_text(size=15),axis.text.y=element_text(size=15))+
  theme(axis.title.x=element_text(size=15),axis.title.y=element_text(size=15))
```

```r
ggplot(df, aes(Age, Profession, col= Work.Experience)) +
  geom_point(col="red", size=3)+
  geom_line(size=1)+
  coord_cartesian(xlim=c(20.0, 50), ylim=c(0, 10))+
  theme(axis.text.x=element_text(size=15),axis.text.y=element_text(size=15))+
  theme(axis.title.x=element_text(size=15),axis.title.y=element_text(size=15))
```
